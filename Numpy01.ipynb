```python
#importing NumPy
import numpy as np
print(np.__version__)
```

```python
#List to NumPy Array
my_np_arr = np.array(
    [1,2,3]
)

print(my_np_arr)
print(type(my_np_arr))
print(my_np_arr.dtype)
print(my_np_arr.shape)
print(my_np_arr.ndim)
```

```python
#Fill an array with a certain size
a = np.zeros(
    shape = (5,)
)
a
```

```python
#Fill an array with a certain value and size and data type
one = np.ones(
    shape = (10,2),
    dtype = 'int'
)
one
```

```python
#Fill an array with a certain value and size
assigned_with_specific_np_array = np.full(
    shape=(10,),
    fill_value = 15,
)
assigned_with_specific_np_array
```

```python
#identity matrix in NumPy
id_matrix = np.eye(4)
print(id_matrix)
```

```python
#Creating a random empty NumPy Array
random_a = np.empty(
    shape=(5,),
    dtype = int
)
random_a
```

```python
#List to NumPy Array
list = [1,2,3]
npArray = np.array(
    list
)
npArray
```

```python
#TypeCast
print(npArray.dtype)
newType = npArray.astype('float')
newType.dtype
```

```python
#np.ar1 = np.arrange(start=1,stop=11, step=1)
ar1 = np.arange(1,11)
ar1
```

```python
#linespace
ar2 = np.linspace(
    start=0,
    stop=10,
    num=9
)
ar2
```

```python
'''
 NumPy Array dimensions
 -scaler ->00
 -vector -->10
 --Matrix -->20
 --Tensor -->xD where x>2

 Tensor is a unit data type of PyTorch and Tensorflow
'''
#Scaler Array
scaler = np.array(47)
#1D Array
oneD = np.array([47, 23, 35, 00])
print(scaler)
print(oneD)
twoD = np.array([[47, 23, 35, 00],[47, 23, 35, 00],[47, 23, 35, 00]])
print(twoD)
print(twoD[1,1])
```

```python
print(twoD[0:2,0:3])
```
