```python
import seaborn as sns
sns.set_style("darkgrid")
sns.__version__
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
tips['day'].value_counts()
```

```python
"Strip Plot"
'''

'''
sns.catplot(data = tips, x = 'day', y ='total_bill', kind = 'strip',
            jitter = False, order = ['Sun', 'Sat', 'Thur', 'Fri'])
```

```python
'''
Swarm Plot

'''
sns.catplot(data = tips, x = 'day', y ='total_bill', kind = 'swarm',
            order = ['Sun', 'Sat', 'Thur', 'Fri'], hue='smoker')
```

```python
sns.catplot(data = tips, x = 'day', y ='total_bill', kind = 'box',
            order = ['Sun', 'Sat', 'Thur', 'Fri'], hue='smoker')
```

```python
'''
Boxen Plot
'''

sns.catplot(
    data = tips,
    x='day',
    y='total_bill',
    kind = 'boxen',
    hue = 'sex'
)
```

```python
'''
Violine Plot : combines box plot and distribution desnity graph...
'''
sns.catplot(
    data = tips,
    x='day',
    y='total_bill',
    kind = 'violin',
    hue = 'smoker'
)
```

```python
'''
count Plot : combines box plot and distribution desnity graph...
'''
sns.catplot(
    data = tips,
    x='day',
    kind = 'count',
    hue = 'smoker'
)
```

```python
'''
Point Plot
'''

titanic= sns.load_dataset("titanic")
titanic.head()
```

```python
sns.catplot(
    data = titanic,
    x='survived',
    y='age',
    kind = 'point',
    hue = 'sex'
)
```
