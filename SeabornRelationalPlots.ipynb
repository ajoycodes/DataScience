```python
import seaborn as sns

sns.set_style(style='darkgrid')
sns.__version__
```

```python
"Data Science"
'''
- Business data
- Data analysis
  - Data cleaning
  - Data preprocessing
  - Data visualization
- Exploratory data analysis: Data analysis with visualization

Relational plots in seaborn:
  Establishes the relationship between two variables/features
  For example, a tabular data containing room size, no of bed rooms, rent
  An example contains the value of each features. [1400, 2, 15k] (Feature Factor)

'''
```

```python
# Sample dataset
tips = sns.load_dataset("tips")

# Basic scatter plot
sns.relplot(data=tips, x="total_bill", y="tip")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="sex")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", style="smoker", size="size")
```

```python
sns.relplot(data=tips, x="size", y="tip", kind="line")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="sex", col="day")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
```

```python
fmri = sns.load_dataset('fmri')
fmri.head()
import matplotlib.pyplot as plt
```

```python
sns.relplot(
    data = fmri,
    x='timepoint',
    y='signal',
    kind='line',
    height = 3,
    col = 'region',
    row = 'event',
    hue = 'region',
    palette = 'muted',
    estimator = None
)
```

```python
sns.relplot(
    data=fmri,
    x='timepoint',
    y='signal',
    kind='line',
    col='subject',
    col_wrap=3,
    hue='region',
    style='event',     # Optional: makes lines dashed/solid by event
    palette='muted',
    linewidth=2.5,
    height=3
)
```
