```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style(style = 'darkgrid')
sns.__version__
```

```python
'''
Statistical Plot:
1. Relational Plot (replot)
2. Distributional Plot (displot)
3. Categorial Plot (catplot)

'''
```

```python
# Sample data
tips = sns.load_dataset("tips")


# Scatter plot showing total bill vs. tip, grouped by gender
sns.relplot(data=tips, x="total_bill", y="tip", hue="sex", kind="scatter")
plt.show()
```

```python
sns.displot(data=tips, x="total_bill", kde=True)
plt.show()
```

```python
sns.catplot(data=tips, x="day", y="tip", kind="box")
plt.show()
```

```python
# Box plot of tips by day of the week
sns.catplot(data=tips, x="day", y="tip", kind="box")
plt.show()
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
sns.displot(data=penguins, x="flipper_length_mm")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", bins=20, kde=True)
```

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack", stat = "probability")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", kind="hist", kde=True)
```

```python
sns.displot(
    data = penguins,
    x = 'species',
    discrete = True
)
```

```python
sns.kdeplot(data=penguins, x="flipper_length_mm", hue="species", )
```

```python
sns.ecdfplot(data=penguins, x="flipper_length_mm")
```

```python
sns.displot(
    data = penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    cbar = True,
    hue = "species"
)
```

```python
g = sns.PairGrid(penguins)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, fill=True)
```
